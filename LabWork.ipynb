# **LAB 3: Simple Flex Lexer**

## **Develop a simple lexer using flex**

The code below is a simple lexer that checks if the given letter is a vowel or not


## **Lex file(vowelz.l)**

```python
%{
	/*To find whether given letter is a vowel or not*/
#undef yywrap
#define yywrap() 1
	void display(int);
%}

%%

[a|e|i|o|u|] {
		int flag=1;
		display(flag);
		return;
		}

.+ {
	int flag=0;
	display(flag);
	return;
	}
	
%%

void display(int flag)
{
	if(flag==1)
		printf("The given letter [%s] is a vowel",yytext);
	else
		printf("The given letter [%s] is NOT a vowel",yytext);		
}

main()
{
	printf("Enter a letter to check if it is a vowel or not");
	yylex(); 
}
```

## **2.a) What is LL (1) parsing?**

A top-down parsing technique used in the syntax analysis stage of compiler design is called LL(1) parsing. The input string, a stack, a parsing table for the given grammar, and a parser are necessary elements for LL(1) parsing.The first L indicates that the input is read from left to right. The second L says that it produces a left-to-right derivation. And the 1 says that it uses one lookahead token.

## **b**) Why do we get the First () and Follow () i.e. what do these two functions really represent? **bold text**

For a given grammar, we must identify FIRST and FOLLOW sets in order for the parser to apply the required rule at the proper place.

We can fill in the blanks of an LL (1) table with the aid of these functions.

In order for the parser to wisely apply the right production rule and produce the proper syntax tree for the supplied input string, we use First.

In order to generate the string from the parse tree, we use Follow to make a Non-terminal vanish out if necessary.

## **Do a quick summary of the front end.**



The phases of a compiler can be assigned to one of three stages.
The stages include :


*   a front end
*  a middle 
*   a backend

The frontend checks the input and verifies syntax and semantics to see whether the program is correctly written in terms of the programming language syntax and semantics.
The frontend generates an intermediate representation of the source code for processing by the middle end.

The middle-end is where optimization of the intermediate representation takes place.Optimization is done by removing useless or unreachable code and reallocation of computation to a less used area
The middle-end also generates an intermediate representation for the backend.

The backend is tasked with translating the intermediate representation from the middle-end to assembly code. The target instruction(s) are chosen for each IR instruction. Here, variables are also selected for the registers.

### **The Front End**


The front-end of a compiler comprises four phases, these are lexical analysis, syntax analysis, and semantic analysis.








image.png





Lexical analysis.
Source code is converted into meaningful lexemes which are represented as tokens by the lexer.

Syntax analysis.
The tokens are input in this phase, they are used to produce a parse tree.

Semantic analysis.
Here we check if the parse tree adheres to the semantics of the programming language used.

Intermediate code generation.
IR is code that is between high and low-level machine code. It is readable and at the same time low level.

Functions of the compiler front-end include;

*   To determine the validity of source code.
*   To determine the content of source code.
*   To make it easier for source code analysis and optimizations.
